```python
import pandas as pd
import scipy.stats as stats

# Categorical Columns to Compare (Removed 'NAFLD' since it caused an error)
categorical_features = ['gender', 'fatigue', 'jaundice', 'digestion', 'skin']

# Statistical results storage
chi2_results = {}

# Perform Chi-Square test for categorical features
for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df['age'].isna())
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    chi2_results[feature] = p

# Convert to DataFrame
chi2_results_df = pd.DataFrame(list(chi2_results.items()), columns=['Feature', 'P-Value'])

# Display the chi-square test results
print("Chi-Square Test Results:\n", chi2_results_df)

# Numerical Columns to Compare
numerical_features = ['AST.sgot', 'ALT.sgpt', 'ALP', 'BIL.T']

# Store p-values for t-test
ttest_results = {}

# Perform T-Test for numerical features
for feature in numerical_features:
    missing_group = df.loc[df['age'].isna(), feature].dropna()
    non_missing_group = df.loc[df['age'].notna(), feature].dropna()

    if len(missing_group) > 1 and len(non_missing_group) > 1:
        t_stat, p_value = stats.ttest_ind(missing_group, non_missing_group, equal_var=False, nan_policy='omit')
        ttest_results[feature] = p_value

# Convert to DataFrame
ttest_results_df = pd.DataFrame(list(ttest_results.items()), columns=['Feature', 'P-Value'])

# Display the T-Test results
print("T-Test Results:\n", ttest_results_df)
```

**HANDLING ALP**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Define medical reference ranges for BIL.T
ALP_NORMAL_MAX = 40  # Upper limit for non-disease (target = 0)
ALP_MILD_MAX = 900    # Upper limit for disease (target = 1)

# Drop rows where predictors are missing (we need AST, ALT, ALP for prediction)
df = df.dropna(subset=['AST.sgot', 'ALT.sgpt', 'NAFLD'])

# Separate rows where BIL.T is missing
train_data = df.dropna(subset=['ALP'])  # Rows where BIL.T is known
test_data = df[df['ALP'].isna()]  # Rows where BIL.T is missing

# Features (X) and target variable (y)
X_train = train_data[['AST.sgot', 'ALT.sgpt', 'NAFLD']]
y_train = train_data['ALP']

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict missing BIL.T values
X_test = test_data[['AST.sgot', 'ALT.sgpt', 'NAFLD']]
predicted_bil_t = model.predict(X_test)

# Apply medical range adjustments
for i, index in enumerate(test_data.index):
    if df.loc[index, 'NAFLD'] == 0:  # Non-disease (Healthy)
        predicted_bil_t[i] = np.clip(predicted_bil_t[i], 0.1, ALP_NORMAL_MAX)
    elif df.loc[index, 'NAFLD'] == 1:  # Disease (Fatty Liver / MASLD)
        predicted_bil_t[i] = np.clip(predicted_bil_t[i], ALP_NORMAL_MAX, ALP_MILD_MAX)

# Fill missing BIL.T values in the original DataFrame
df.loc[df['ALP'].isna(), 'ALP'] = predicted_bil_t

# Print confirmation
print("Missing ALP values have been filled using regression with medical range adjustments.")
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, _tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    recall_score,
    precision_score,
    f1_score,
    classification_report,
)
from imblearn.over_sampling import SMOTE

# Define feature set
features = ['gender_1.0', 'fatigue_1.0', 'jaundice_1.0', 'digestion_1.0', 'skin_1.0',
            'ALP', 'BIL.T', 'ALT.sgpt', 'AST.sgot', 'age']

X = df_normalized[features]
y = df_normalized["NAFLD"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Train initial decision tree to find top features
dt_clf = DecisionTreeClassifier(max_depth=3, min_samples_leaf=2, random_state=42, class_weight='balanced')
dt_clf.fit(X_train, y_train)

# Select top 4 features
feature_importances = dt_clf.feature_importances_
important_features = sorted(zip(X_train.columns, feature_importances), key=lambda x: x[1], reverse=True)[:4]
top_features = [feature[0] for feature in important_features]
print("Top 4 most important features:", top_features)

# Train decision tree on top features
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]

dt_selected = DecisionTreeClassifier(max_depth=3, min_samples_leaf=2, random_state=42, class_weight='balanced')
dt_selected.fit(X_train_selected, y_train)

# Predict
y_pred = dt_selected.predict(X_test_selected)

# Get leaf indices for test samples
leaf_indices = dt_selected.apply(X_test_selected)
tree = dt_selected.tree_

# Helper functions to extract rules
def _get_leaves(tree, node_id):
    if tree.children_left[node_id] == _tree.TREE_LEAF:
        return [node_id]
    left_leaves = _get_leaves(tree, tree.children_left[node_id])
    right_leaves = _get_leaves(tree, tree.children_right[node_id])
    return left_leaves + right_leaves

def get_rule_path(leaf_id, feature_names):
    path = []
    node = 0
    while node != leaf_id:
        left = tree.children_left[node]
        right = tree.children_right[node]
        threshold = tree.threshold[node]
        feature = feature_names[tree.feature[node]]

        if leaf_id in _get_leaves(tree, left):
            path.append(f"{feature} <= {threshold:.2f}")
            node = left
        else:
            path.append(f"{feature} > {threshold:.2f}")
            node = right
    return " AND ".join(path)

# Rule-wise PPV and NPV calculation
print("\n=== Decision Tree Rules with PPV & NPV ===")

for leaf in np.unique(leaf_indices):
    indices = np.where(leaf_indices == leaf)[0]
    y_true_leaf = y_test.iloc[indices]
    y_pred_leaf = y_pred[indices]

    if len(y_true_leaf) == 0:
        continue

    cm_leaf = confusion_matrix(y_true_leaf, y_pred_leaf, labels=[0, 1])
    if cm_leaf.shape != (2, 2):
        continue

    TN, FP, FN, TP = cm_leaf.ravel()
    support_0 = TN + FP
    support_1 = TP + FN
    ppv_1 = TP / (TP + FP) if (TP + FP) != 0 else 0
    npv_0 = TN / (TN + FN) if (TN + FN) != 0 else 0

    rule = get_rule_path(leaf, top_features)

    # Extract the classification result from the rule
    predicted_class = "1.0" if TP + FP > FN + TN else "0.0"

    # Print the rule with PPV & NPV next to the class
    print(f"\nRule: IF {rule}")
    print(f"  -> class: {predicted_class} | PPV: {ppv_1:.2f} | NPV: {npv_0:.2f}")




# Train a new Decision Tree using only the selected features
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]

dt_selected = DecisionTreeClassifier(max_depth=3, min_samples_leaf=2, random_state=42, class_weight='balanced')
dt_selected.fit(X_train_selected, y_train)

# Extract rules from the decision tree
rules_text = export_text(dt_selected, feature_names=top_features)

# Make predictions
y_pred = dt_selected.predict(X_test_selected)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Extract TP, FN, FP, TN (only valid for binary classification)
if len(np.unique(y_test)) == 2:
    TP, FN, FP, TN = cm[1, 1], cm[1, 0], cm[0, 1], cm[0, 0]

    # Compute evaluation metrics using sklearn
    PPV = precision_score(y_test, y_pred, pos_label=1)  # Positive Predictive Value
    NPV = TN / (TN + FN) if (TN + FN) != 0 else 0  # Negative Predictive Value
    Accuracy = accuracy_score(y_test, y_pred)  # Overall accuracy
    Recall = recall_score(y_test, y_pred)  # Sensitivity (Recall)
    Specificity = TN / (TN + FP) if (TN + FP) != 0 else 0  # True Negative Rate
    F1 = f1_score(y_test, y_pred)  # F1-Score

    # Compute support values
    support_yes = TP + FN  # Support for class Yes
    support_no = TN + FP   # Support for class No

    # Print results in a structured format
    print("\n=== Model Performance ===")
    print(f"Accuracy: {Accuracy:.2f}")
    print(f"PPV (Positive Predictive Value): {PPV:.2f}")
    print(f"NPV (Negative Predictive Value): {NPV:.2f}")
    print(f"Recall (Sensitivity): {Recall:.2f}")
    print(f"Specificity (True Negative Rate): {Specificity:.2f}")
    print(f"F1-Score: {F1:.2f}")
    print(f"Support for Yes class: {support_yes}")
    print(f"Support for No class: {support_no}")
```
