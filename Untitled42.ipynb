```python
import pandas as pd
import scipy.stats as stats

# Categorical Columns to Compare (Removed 'NAFLD' since it caused an error)
categorical_features = ['gender', 'fatigue', 'jaundice', 'digestion', 'skin']

# Statistical results storage
chi2_results = {}

# Perform Chi-Square test for categorical features
for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df['age'].isna())
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    chi2_results[feature] = p

# Convert to DataFrame
chi2_results_df = pd.DataFrame(list(chi2_results.items()), columns=['Feature', 'P-Value'])

# Display the chi-square test results
print("Chi-Square Test Results:\n", chi2_results_df)

# Numerical Columns to Compare
numerical_features = ['AST.sgot', 'ALT.sgpt', 'ALP', 'BIL.T']

# Store p-values for t-test
ttest_results = {}

# Perform T-Test for numerical features
for feature in numerical_features:
    missing_group = df.loc[df['age'].isna(), feature].dropna()
    non_missing_group = df.loc[df['age'].notna(), feature].dropna()

    if len(missing_group) > 1 and len(non_missing_group) > 1:
        t_stat, p_value = stats.ttest_ind(missing_group, non_missing_group, equal_var=False, nan_policy='omit')
        ttest_results[feature] = p_value

# Convert to DataFrame
ttest_results_df = pd.DataFrame(list(ttest_results.items()), columns=['Feature', 'P-Value'])

# Display the T-Test results
print("T-Test Results:\n", ttest_results_df)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Define medical reference ranges for BIL.T
BIL_T_NORMAL_MAX = 1.2  # Upper limit for non-disease (target = 0)
BIL_T_MILD_MAX = 3.0    # Upper limit for disease (target = 1)

# Drop rows where predictors are missing (we need AST, ALT, ALP for prediction)
df = df.dropna(subset=['AST.sgot', 'ALT.sgpt', 'ALP', 'NAFLD'])

# Separate rows where BIL.T is missing
train_data = df.dropna(subset=['BIL.T'])  # Rows where BIL.T is known
test_data = df[df['BIL.T'].isna()]  # Rows where BIL.T is missing

# Features (X) and target variable (y)
X_train = train_data[['AST.sgot', 'ALT.sgpt', 'ALP', 'NAFLD']]
y_train = train_data['BIL.T']

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict missing BIL.T values
X_test = test_data[['AST.sgot', 'ALT.sgpt', 'ALP', 'NAFLD']]
predicted_bil_t = model.predict(X_test).round(2)

# Apply medical range adjustments
for i, index in enumerate(test_data.index):
    if df.loc[index, 'NAFLD'] == 0:  # Non-disease (Healthy)
        predicted_bil_t[i] = np.clip(predicted_bil_t[i], 0.1, BIL_T_NORMAL_MAX)
    elif df.loc[index, 'NAFLD'] == 1:  # Disease (Fatty Liver / MASLD)
        predicted_bil_t[i] = np.clip(predicted_bil_t[i], BIL_T_NORMAL_MAX, BIL_T_MILD_MAX)

# Fill missing BIL.T values in the original DataFrame
df.loc[df['BIL.T'].isna(), 'BIL.T'] = predicted_bil_t

# Print confirmation
print("Missing BIL.T values have been filled using regression with medical range adjustments.")
```
